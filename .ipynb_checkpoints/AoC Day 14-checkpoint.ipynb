## First Attempt

```python
from pathlib import Path
import numpy as np
```

```python
dtype = 'string'
txt=[]

f_in = Path('./Data Sets/day14data.txt')
with open(f_in,'r') as f:
    txt = f.read().split('\n')

if(dtype == 'int'):
    nums = []
    for num in txt:
        nums.append(int(num))
elif(dtype == 'float'):
    nums = []
    for num in txt:
        nums.append(float(num))
print(txt)
# print(nums)

start = []
dxs = []
dys = []

xs = set()
ys = set()

coords = []
for line in txt:
    cs = line.split(' -> ')
    dxcs = []
    dycs = []
    for idx, c in enumerate(cs):
        x,y = c.split(',')
        x = int(x)
        y = int(y)
        print(x, y)
        xs.add(x)
        ys.add(y)
        
        if idx == 0:
            start.append((x,y))
            cx = x
            cy = y
        else:
            dx = x - cx
            dy = y - cy
            cx = x
            cy = y
            dxcs.append(dx)
            dycs.append(dy)
    dxs.append(dxcs)
    dys.append(dycs)
```

```python
print(min(xs), max(xs))
print(min(ys), max(ys))

xrange = max(xs) - min(xs) + 3
yrange = (max(ys) - min(ys))*5
print(xrange, yrange)
```

```python
grid = np.zeros((xrange,yrange))
grid.T
```

```python
grid = np.zeros((xrange,yrange))

# loops through all the lines
for idx, c in enumerate(start):
    dx_list = dxs[idx]
    dy_list = dys[idx]
    
    cx, cy = c
    # loops through all segments in the lines
    for jdx, dx in enumerate(dx_list):
        dy = dy_list[jdx]
#         print(dx,dy)
        # loops through all the points to be set
        for delta in range(max(abs(dx),abs(dy))+1):
            if abs(dx) > 0:
                grid[cx + delta*np.sign(dx) - min(xs)+1, grid.shape[1] - max(ys) + (cy - 1)] = 1
#                 print(cx + delta - min(xs), cy - min(ys))
            else:
                grid[cx - min(xs)+1, grid.shape[1] - max(ys) + (cy + delta*np.sign(dy) - 1)] = 1
#                 print(cx - min(xs), cy + delta - min(ys))
        cx = cx + dx
        cy = cy + dy
```

```python
drop_point = 500 - min(xs) + 1

def fall_step(grid, xdx, ydx):
    if grid[xdx,ydx+1] == 0:
        return (xdx, ydx+1)
    elif grid[xdx-1,ydx+1] == 0:
        return (xdx-1, ydx+1)
    elif grid[xdx+1, ydx+1] == 0:
        return (xdx+1, ydx+1)
    else:
        return(xdx, ydx)
    
def sand_drop(grid, start):
    x, y = start
    
    still_falling = True
    while still_falling:
#         print(x,y)
        nx, ny = fall_step(grid, x, y)
        if nx == x and ny == y:
            still_falling = False
            grid[nx, ny] = 2
            return False
        elif ny == yrange-1:
            return True
        else:
            x = nx
            y = ny

def how_much_sand(grid, start):
    count = 0
    sand_stopped = False
    while not sand_stopped:
        sand_stopped = sand_drop(grid, start)
        count +=1
    return count-1
    
    
            
```

```python
xrange = (max(xs))*2
yrange = (max(ys))+3
print(xrange, yrange)

grid = np.zeros((xrange,yrange),dtype=int)

# loops through all the lines
for idx, c in enumerate(start):
    dx_list = dxs[idx]
    dy_list = dys[idx]
    
    cx, cy = c
    # loops through all segments in the lines
    for jdx, dx in enumerate(dx_list):
        dy = dy_list[jdx]
        # loops through all the points to be set
        for delta in range(max(abs(dx),abs(dy))+1):
            if abs(dx) > 0:
                grid[cx + delta*np.sign(dx), cy] = 1
            else:
                grid[cx, cy + delta*np.sign(dy)] = 1
        cx = cx + dx
        cy = cy + dy
grid[:,grid.shape[1]-1] = 3
```

```python
def fall_step(grid, xdx, ydx):
    if grid[xdx,ydx+1] == 0:
        return (xdx, ydx+1)
    elif grid[xdx-1,ydx+1] == 0:
        return (xdx-1, ydx+1)
    elif grid[xdx+1, ydx+1] == 0:
        return (xdx+1, ydx+1)
    else:
        return(xdx, ydx)
    
def sand_drop(grid, start):
    x, y = start
    
    still_falling = True
    while still_falling:
#         print(x,y)
        nx, ny = fall_step(grid, x, y)
        if ny == 0:
            return True
        elif nx == x and ny == y:
            still_falling = False
            grid[nx, ny] = 2
            return False
        else:
            x = nx
            y = ny

def how_much_sand(grid, start):
    count = 0
    sand_stopped = False
    while not sand_stopped:
        sand_stopped = sand_drop(grid, start)
        count +=1
    return count         
```

```python
how_much_sand(grid, (500, 0))
```
