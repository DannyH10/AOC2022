```python
from pathlib import Path
```

```python
dtype = 'string'
txt=[]

f_in = Path('./Data Sets/day5data.txt')
with open(f_in,'r') as f:
    boxes, inst = f.read().split('\n\n')
b = boxes.split('\n')
print(b)
print(inst)
inst = inst.split('\n')
```

## Part 1

```python
stacks = [int(a) for a in b.pop().strip().split('   ')]

print(stacks)
# print(b)

storage = []

for _ in stacks:
    storage.append(list())

print(storage)
    
for row in b[::-1]:
    values = list(row[1::4])
    for idx, l in enumerate(storage):
        if values[idx] != ' ':
            l.append(values[idx])
print(values)
print(storage)
```

```python
import regex as re
```

```python
parsed_i = []

for i in inst:
    num_to_move, start, end = re.findall('[0-9]+', i)
    parsed_i.append((int(num_to_move), int(start), int(end)))
parsed_i
```

```python
def move_box(storage, start, end):
    box = storage[start-1].pop()
    storage[end-1].append(box)
    
for i in parsed_i:
    moves, start, end = i
    for _ in range(moves):
        move_box(storage, start, end)

storage
```

```python
message = ''

for s in storage:
    message += s.pop()
    
print(message)
```

```python
dtype = 'string'
txt=[]

f_in = Path('./Data Sets/day5data.txt')
with open(f_in,'r') as f:
    boxes, inst = f.read().split('\n\n')
b = boxes.split('\n')
print(b)
print(inst)
inst = inst.split('\n')
```

```python
stacks = [int(a) for a in b.pop().strip().split('   ')]

print(stacks)
# print(b)

storage = []

for _ in stacks:
    storage.append(list())

print(storage)
    
for row in b[::-1]:
    values = list(row[1::4])
    for idx, l in enumerate(storage):
        if values[idx] != ' ':
            l.append(values[idx])
print(values)
print(storage)
```

```python
import regex as re
```

```python
parsed_i = []

for i in inst:
    num_to_move, start, end = re.findall('[0-9]+', i)
    parsed_i.append((int(num_to_move), int(start), int(end)))
parsed_i
```

```python
def move_boxes(storage, moves, start, end):
    temp = list()
    for _ in range(moves):
        temp.append(storage[start-1].pop())
    
    for val in temp[::-1]:
        storage[end-1].append(val)
    
for i in parsed_i:
    moves, start, end = i
    move_boxes(storage, moves, start, end)

storage
```

```python
message = ''

for s in storage:
    message += s.pop()
    
print(message)
```
