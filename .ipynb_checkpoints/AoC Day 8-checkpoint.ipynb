## First Attempt

```python
from pathlib import Path
```

```python
dtype = 'int'
txt=[]

f_in = Path('./Data Sets/day8data.txt')
with open(f_in,'r') as f:
    txt = f.read().split('\n')

if(dtype == 'int'):
    nums = []
    for num in txt:
        temp = list(num)
        nums.append(temp)
elif(dtype == 'float'):
    nums = []
    for num in txt:
        nums.append(float(num))
print(nums)
```

## Part 1

```python
import numpy as np
```

```python
trees = np.array(nums, dtype=int)
```

```python
def is_hidden(f, xdx, ydx):
    l_hidden = np.max(f[:xdx,ydx]) >= f[xdx,ydx]
    r_hidden = np.max(f[xdx+1:,ydx]) >= f[xdx,ydx]
    t_hidden = np.max(f[xdx,:ydx]) >= f[xdx,ydx]
    b_hidden = np.max(f[xdx,ydx+1:]) >= f[xdx,ydx]
    return l_hidden and r_hidden and t_hidden and b_hidden
```

```python
can_be_seen = np.ones(trees.shape, dtype=bool)

rows, columns = trees.shape
for j in range(columns):
    for i in range(rows):
        if(i == 0 or j==0 or i == rows - 1 or j == columns - 1):
            continue
        else:
            can_be_seen[i,j] = not is_hidden(trees, i, j)
```

```python
print(np.sum(can_be_seen))
```

## Part 2

```python
def scenic_score(f, xdx, ydx):
    l_prod = (f[:xdx,ydx] < f[xdx, ydx])[::-1]
    l_num = np.argmin(l_prod) + 1
    r_prod = f[xdx+1:,ydx] < f[xdx,ydx]
    r_num = np.argmin(r_prod) + 1
    t_prod = (f[xdx,:ydx] < f[xdx,ydx])[::-1]
    t_num = np.argmin(t_prod) + 1
    b_prod = f[xdx,ydx+1:] < f[xdx, ydx]

    b_num = np.argmin(b_prod) + 1

    if(np.all(b_prod)):
        b_num = b_prod.shape[0]
#         print(b_num)
    if(np.all(t_prod)):
        t_num = t_prod.shape[0]
#         print(t_num)
    if(np.all(l_prod)):
        l_num = l_prod.shape[0]
#         print(l_num)
    if(np.all(r_prod)):
        r_num = r_prod.shape[0]
#         print(r_num)
    return l_num * r_num * t_num * b_num
```

```python
scenic_score(trees, 1,2)
```

```python
can_be_seen = np.ones(trees.shape, dtype=bool)

max_score = 0

rows, columns = trees.shape
for j in range(columns):
    for i in range(rows):
        if(i == 0 or j==0 or i == rows - 1 or j == columns - 1):
            continue
        if(can_be_seen[i,j]):
            score = scenic_score(trees, i, j)
            if(score > max_score):
                max_score = score
print(max_score)
```
