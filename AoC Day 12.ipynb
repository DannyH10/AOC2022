```python
from pathlib import Path
import numpy as np
import networkx as nx
```

```python
dtype = 'string'
txt=[]

f_in = Path('./Data Sets/day12data.txt')
with open(f_in,'r') as f:
    txt = f.read().split('\n')

if(dtype == 'int'):
    nums = []
    for num in txt:
        nums.append(int(num))
elif(dtype == 'float'):
    nums = []
    for num in txt:
        nums.append(float(num))
print(txt)
# print(nums)
grid = []

for line in txt:
    grid.append(list(line))
    
grid_array = np.array(grid)
g = np.zeros(grid_array.shape, dtype=int)

for idx in range(grid_array.shape[0]):
    for jdx in range(grid_array.shape[1]):
        g[idx,jdx] = ord(grid_array[idx, jdx])
g = g - ord('a')
g
```

```python
x_max = g.shape[0]-1
y_max = g.shape[1]-1

print(x_max)
print(y_max)

def possible_steps(loc):
    x, y = loc
    coords = [(x-1,y),(x,y+1),(x+1,y),(x,y-1)]
    if x == x_max:
        coords.remove((x+1,y))
    elif x == 0:
        coords.remove((x-1,y))
    if y == y_max:
        coords.remove((x,y+1))
    elif y == 0:
        coords.remove((x,y-1))
        
    return coords
    
possible_steps(np.array([0,2]))
```

```python
m_range = nx.DiGraph()

letter_count = {}
a_squares = set()

for idx in range(grid_array.shape[0]):
    for jdx in range(grid_array.shape[1]):
        # add to the counters for the indices in the graph
        letter = grid_array[idx, jdx]
        if letter not in list(letter_count.keys()):
            letter_count[letter] = 0
        else:
            letter_count[letter] += 1
            
        # add the node to the graph
        if letter == 'S':
            node = 'Start'
        elif letter == 'E':
            node = 'End'
        else:
            node = letter + '_' + str(idx) + '_' + str(jdx)
        m_range.add_node(node)
        
        if letter == 'a':
            a_squares.add(node)
        
        # make connections
        neighbors = possible_steps((idx,jdx))
        if letter == 'S':
            letter_ascii = ord('a')
        elif letter == 'E':
            letter_ascii = ord('z')
        else:
            letter_ascii = ord(grid_array[idx, jdx])
#         print(idx,jdx)
#         print(neighbors)
        for n in neighbors:
            n_letter = grid_array[n[0],n[1]]
            if n_letter == 'S':
                n_ascii = ord('a')
                n_node = 'Start'
            elif n_letter == 'E':
                n_ascii = ord('z')
                n_node = 'End'
            else:
                n_ascii = ord(n_letter)
                n_node = n_letter + '_' + str(n[0]) + '_' + str(n[1])
            if n_ascii - letter_ascii < 1.5:
                m_range.add_node(n_node)
                m_range.add_edge(node, n_node)
        
# nx.draw(m_range, with_labels='True')       
```

```python
len(nx.dijkstra_path(m_range, 'Start', 'End'))-1
```
