```python
def eval_expression(exp):
    exp = exp.replace(' ', '', len(exp))
    explist = list(exp)
    parenthesis = exp.count('(')
    while(parenthesis > 0):
        start = len(explist) - explist[::-1].index('(', 0, len(explist)) - 1
        end = explist.index(')', start)
        inside = explist[start+1:end]
        sub_total = eval_cluster(inside)
        explist = explist[:start] + [str(sub_total)] + explist[end+1:]
        parenthesis -=1
    return eval_cluster(explist)

def eval_cluster(cluster):
    total = int(cluster[0])
    for idx in range(1,len(cluster),2):
        operator = cluster[idx]
        num = cluster[idx+1]
        if(operator == '+'):
            total += int(num)
        else:
            total *= int(num)
    return total

def sum_eqns(list_to_sum):
    summation = 0
    for exp in list_to_sum:
        summation += eval_expression(exp)
    return summation
```
