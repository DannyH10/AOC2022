```python
def parse_input(entry):
    rulelist = entry[0].split('\n')
    messages = entry[1].split('\n')
    
    rules = {}
    for rule in rulelist:
        key,val = rule.split(':')
        rs = []
        print(val)
        print(val[2])
        if not val[2].isalpha():
            val = val.split('|')
            for v in val:
                total_entries = []
                nums = v.split()
                intlist = []
                for num in nums:
                    intlist.append(int(num))
                total_entries.append(intlist)
                rs.append(intlist)
            rules[int(key)] = rs
        else:
            rules[int(key)] = val[2]
    return [messages, rules]
```
