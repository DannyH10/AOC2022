```python
def parse_entries(cell):
    cell_num,image = cell.split(':')
    cell_num = int(cell_num.split()[1])
    image = image[1:].split('\n')
    image_list = []
    for entry in image:
        image_list.append(list(entry))
    return [cell_num, image_list]

def parse_file(file):
    cells = []
    cell_maps = {}
    for cell in file:
        id_num, image = parse_entries(cell)
        cells.append(id_num)
        cell_maps[id_num] = image
    return [cells, cell_maps]
```

```python
def get_edges(cell):
    edges = []
    
    # top and bottom
    edges.append(cell[0])
    edges.append(cell[-1])
    
    # sides
    left = []
    right = []
#     print(cell)
    for row in cell:
#         print(row)
        left.append(row[0])
        right.append(row[-1])
    edges.append(left)
    edges.append(right)
    
    return edges

def find_all_edges(idnums, cells):
    cell_to_edge = {}
    for idnum in idnums:
        cell_to_edge[idnum] = get_edges(cells[idnum])
    return cell_to_edge

def find_corners(file):
    ids, cells = parse_file(file)
    edge_map = find_all_edges(ids, cells)
#     print(ids)
#     print(cells)
    adjacent_cells = defaultdict(set)
    for idx in ids:
        checklist = ids.copy()
        checklist.remove(idx)
        for check in checklist:
            for edge in edge_map[idx]:
                if(edge in edge_map[check] or edge[::-1] in edge_map[check]):
                    adjacent_cells[check].add(idx)
                    adjacent_cells[idx].add(check)
    corner_product = 1
    for idx in ids:
        if len(adjacent_cells[idx]) == 2:
            corner_product *= idx
    print(corner_product)
    return adjacent_cells
```

```python
def get_top_row(adjacent_cells, corner, lsize):
    # setup top row
    row = []
    row.append(corner)
    current = corner
    for _ in range(lsize - 2):
        adj = adjacent_cells[current]
        for a in adj:
            if len(adjacent_cells[a]) ==3:
                row.append(a)
                current = a
                break
    adj = adjacent_cells[current] 
    for a in adj:
        if len(adjacent_cells[a]) ==2:
            row.append(a)
            current = a
            break
    return row

def get_middle_row(adjacent_cells, top_row, lsize, first):
    row = []
    row.append(first)
    current = first
    for idx in range(1,lsize):
        next_cell = set()
        adj_left = set(adjacent_cells[current])
        adj_top = set(adjacent_cells[top_row[idx]])
        cells = adj_left.intersection(adj_top)
        for cell in cells:
            if(cell not in top_row):
                row.append(cell)
                break
        current = cell
    return row

def get_unused_cell(used_cells, cell_list):
    for cell in cell_list:
        if(cell not in used_cells):
            return cell

def organize_cells(adjacent_cells, corner, lsize):
    vals_used = []
    image = []
    row = get_top_row(adjacent_cells, corner, lsize)
    vals_used += row
    image.append(row)
    
    for rowdx in range(1, lsize):
        adj_prev_first = adjacent_cells[row[0]]
        first = get_unused_cell(vals_used, adj_prev_first)
        row = get_middle_row(adjacent_cells, row, lsize, first)
        image.append(row)
        vals_used += row
    return image
```
