```python
def get_tile_flipped(tile_input):
    formated_input = tile_input.replace('e', 'e ', len(tile_input))
    formated_input = formated_input.replace('w', 'w ', len(tile_input))
    direction_list = formated_input[:-1].split(' ')
    
    # put the directions in a dictionary
    direct_dict = {}
    possible_directions = ['e', 'w', 'se', 'sw', 'nw', 'ne']
    for d in possible_directions:
        direct_dict[d] = direction_list.count(d)
    se = direct_dict['se']
    nw = direct_dict['nw']

    sw = direct_dict['sw']
    ne = direct_dict['ne']
    
    e = direct_dict['e']
    w = direct_dict['w']
    
    tile_name = []
    x = 0
    y = 0
    if(se > nw):
        tile_name.append(str(se - nw) + 'se')
        x += (se-nw)/2
        y += -(se-nw)/2
    else:
        tile_name.append(str(nw - se) + 'nw')
        x += (se-nw)/2
        y += -(se-nw)/2
        
    if(sw > ne):
        tile_name.append(str(sw - ne) + 'sw')
        x += -(sw-ne)/2
        y += -(sw-ne)/2
    else:
        tile_name.append(str(ne - sw) + 'ne')
        x += -(sw-ne)/2
        y += -(sw-ne)/2#((2)**(1/2))
        
    if(e > w):
        tile_name.append(str(e - w) + 'e')
        x += e-w
    else:
        tile_name.append(str(w - e) + 'w')
        x += e-w
        
    return (x, y)
```

```python
def count_tiles(file):
    tile_summary = {}
    black_tiles = []
    for tile in file:
        tile_name = get_tile_flipped(tile)
        if(tile_name in tile_summary.keys()):
            tile_summary[tile_name] += 1
        else:
            tile_summary[tile_name] = 1
            
    black = 0
    white = 0
    for key in tile_summary.keys():
        if(tile_summary[key] % 2 == 0):
            white +=1
        else:
            black +=1
            black_tiles.append(key)
    return [black, white, tile_summary, black_tiles]
```

```python
def get_adj_tiles(tile):
    x,y = tile
    return [(x-1,y), (x+1,y), (x-0.5,y-0.5), (x-0.5, y + 0.5), (x + 0.5,y + 0.5), (x + 0.5,y - 0.5)]

def tiles_to_check(black_tiles):
    tiles_to_check = []
    for tile in black_tiles:
        tiles_to_check += get_adj_tiles(tile)
    return tiles_to_check

def get_black_tiles(black_tiles):
    check_these = set(tiles_to_check(black_tiles))
    blacks = set(black_tiles)
    whites = check_these.difference(blacks)
    
    new_blacks = set()
    for b in blacks:
        adj = set(get_adj_tiles(b))
        intersection = adj.intersection(blacks)
        num = len(intersection)
        if(num == 1 or num == 2):
            new_blacks.add(b)
    for w in whites:
        adj = set(get_adj_tiles(w))
        intersection = adj.intersection(blacks)
        num = len(intersection)
        if(num == 2):
            new_blacks.add(w)
    return new_blacks

def flip_tiles_for(days, black_tiles):
    for _ in range(days):
        black_tiles = get_black_tiles(black_tiles)
    return len(black_tiles)
```
