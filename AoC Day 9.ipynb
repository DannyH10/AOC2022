```python
head = (0, 0)
tail = (0, 0)
history = set()

def distance(p0, p1):
    return math.sqrt((p0[0] - p1[0])**2 + (p0[1] - p1[1])**2)

def move_tail(direc, head, tail):
    if distance(head, tail) > 1.5:
        if direc == 'R':
            tail = (head[0] - 1, head[1])
        elif direc == 'L':
            tail = (head[0] + 1, head[1])
        elif direc == 'U':
            tail = (head[0], head[1] - 1)
        else:
            tail = (head[0], head[1] + 1)    
    return tail

for line in txt:
    direc, step = line.split(' ')
    for s in range(int(step)):
        if direc == 'R':
            head = (head[0] + 1, head[1])
        elif direc == 'L':
            head = (head[0] - 1, head[1])
        elif direc == 'U':
            head = (head[0], head[1] + 1)
        else:
            head = (head[0], head[1] - 1)
        tail = move_tail(direc, head, tail)
        history.add(tail)

print(len(history))
```

```python
knots = [(0,0) for _ in range(10)]
history = set()

def distance(p0, p1):
    return math.sqrt((p0[0] - p1[0])**2 + (p0[1] - p1[1])**2)

def move_tail(direc, head, tail):
    if distance(head, tail) > 2.5:
        tail = (int((head[0] + tail[0])/2),int((head[1] + tail[1])/2))
    if distance(head, tail) > 1.5:
        if abs(tail[0] - head[0]) > 1.5:
            tail = (int((head[0] + tail[0])/2), head[1])
        else:
            tail = (head[0], int((head[1] + tail[1])/2))
    return tail

for line in txt:
    direc, step = line.split(' ')
    for s in range(int(step)):
        head = knots[0]
        if direc == 'R':
            newhead = (head[0] + 1, head[1])
        elif direc == 'L':
            newhead = (head[0] - 1, head[1])
        elif direc == 'U':
            newhead = (head[0], head[1] + 1)
        else:
            newhead = (head[0], head[1] - 1)
        knots[0] = newhead
        for idx in range(len(knots)-1):
            knots[idx + 1] = move_tail(direc, knots[idx], knots[idx + 1])
        history.add(knots[len(knots)-1])
#         for idx, k in enumerate(knots):
#             print('Knot -',idx, '--', k)
print(len(history))
```
